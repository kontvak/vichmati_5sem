__Упражнение 4 -  Погрешность определения корня уравнения__

С каким числом верных знаков (или относительной погрешностью) должен быть известен свободный член в уравнении $x^2-2 x+0.999993751=0$, чтобы корни имели четыре верных знака?

Одно из решений $1 + (1-c)^{0.5}$. Тогда $\Delta x = \Delta c / 2 / (1-c)^{0.5}$. При значении $\Delta x = 0.5 \cdot 10^{-4}$ значение $\Delta c = 2.5\cdot 10^{-7}$.

```python
const_c = 0.999993751
delta_const_c = 2.5*10**(-7)
def first_solution(c):
    return 1 + (1-c)**0.5

print(f"delta_const_c = {delta_const_c} -->\ndelta_c = {abs(first_solution(const_c + delta_const_c) - first_solution(const_c))}")
```

__Упражнение 5 -  Рост погрешности в последовательности__

Пусть задана последовательность чисел $x_n, n=0,1,2, \ldots$, причем $5 x_{n+1}-x_n=4$, а $x_0$ известно с относительной погрешностью $10^{-6}$. При каких значениях $x_0$ относительная погрешность при вычислении $x_n$ будет быстро возрастать с ростом $n$ ?

Линейное уравнение, значит общее однородное $5x_{n+1}-x_{n}=0$ имеет решение $x_n=const(1/5)^n$. Частное решение $x_n=1$. Тогда решением является $x_n=const(1/5)^n+1$, и подставляя ноль, находим константу $x_0 = const+1$. Решение : $x_n=(x_0-1)(1/5)^n+1$. Погрешности :
$$\Delta x_n / \Delta x_0 = dx_n/dx_0 = 0.2^n$$
$$\Delta x_n / x_n = (0.2^n*\Delta x_0)/((x_0-1)*0.2^n+1)$$
$$\Delta x_n / x_n = \Delta x_0/(5^n+(x_0-1))$$
Из асимптотики видно, что относительная погрешность $x_n$ убывает, вне зависимости от $x_0$.


__Упражнение 5.2 -  Рост погрешности в последовательности__

Пусть задана последовательность чисел $x_n, n=0,1,2, \ldots$, причем $x_{n+1}-5 x_n=4$, а $x_0$ известно с относительной погрешностью $10^{-6}$. При каких значениях $x_0$ относительная погрешность при вычислении $x_n$ будет быстро возрастать с ростом $n$ ?

Линейное уравнение, значит общее однородное $x_{n+1}-5x_{n}=0$ имеет решение $x_n=const(5)^n$. Частное решение $x_n=-1$. Тогда решением является $x_n=const*5^n-1$, и подставляя ноль, находим константу $x_0 = const-1$. Решение: $x_n=(x_0+1)5^n-1$. Погрешности:
$$
\Delta x_n/x_n =| 5^n \Delta x_0 /((x_0+1)5^n-1) |= |5^n 10^{-6} x_0 /((x_0+1)5^n-1)|.
$$
При $x_0 \neq -1$:
$$\Delta x_n/x_n \approx =| 10^{-6} x_0 /(x_0+1) |= const.$$
При $x_0 = -1$:
$$\Delta x_n/x_n = |5^n 10^{-6} x_0 |.$$
То есть при $x_0 = -1$ рост экспоненциальный.
